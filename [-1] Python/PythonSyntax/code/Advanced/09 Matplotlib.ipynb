【例】随机漫步

```python
from random import choice

class RandomWalk():
    """一个生产随机漫步的类"""
    def __init__(self, num_points=5000):
        self.num_points = num_points
        self.x_values = [0]
        self.y_values = [0]
    
    def fill_walk(self):
        while len(self.x_values) < self.num_points:
            x_direction = choice([1, -1])
            x_distance = choice([0, 1, 2, 3, 4])
            x_step = x_direction * x_distance
            
#             - 随机选择x方向的移动方向和距离，并计算x的步长（x_step）
#             - 随机选择y方向的移动方向和距离，并计算y的步长（y_step）
#             - 如果x_step和y_step都为0，表示未移动，则跳过本次循环
#             - 计算下一个点的x坐标和y坐标，并将其添加到相应的列表中
    
            y_direction = choice([1, -1])
            y_distance = choice([0, 1, 2, 3, 4])
            y_step = y_direction * y_distance            
        
            if x_step == 0 or y_step == 0:
                continue
            next_x = self.x_values[-1] + x_step
            next_y = self.y_values[-1] + y_step
            self.x_values.append(next_x)
            self.y_values.append(next_y)
```

```python
rw = RandomWalk(10000)
rw.fill_walk()
point_numbers = list(range(rw.num_points))
plt.figure(figsize=(12, 6))                 
plt.scatter(rw.x_values, rw.y_values, c=point_numbers, cmap="inferno", s=1)
plt.colorbar()
plt.scatter(0, 0, c="green", s=100)
plt.scatter(rw.x_values[-1], rw.y_values[-1], c="red", s=100)

plt.xticks([])
plt.yticks([])

# 使用plt.figure创建一个图形窗口，设置图形尺寸为12x6。
# 
# 使用plt.scatter绘制散点图，传入rw.x_values和rw.y_values作为数据，c参数设置点的颜色，取值为point_numbers，cmap参数设置颜色的映射，s参数设置点的大小为1。
# 
# 使用plt.colorbar添加一个颜色条，用于对应点的颜色和值。
# 
# 使用plt.scatter再次绘制两个点，分别表示起点和终点，起点使用绿色表示，终点使用红色表示。
# 
# 使用plt.xticks([])和plt.yticks([])取消x轴和y轴的刻度标签。
```

### 13.1.4 多子图

【1】简单多子图

```python
def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)

t1 = np.arange(0.0, 5.0, 0.1)
# 使用np.arange函数生成一个从0到5的步长为0.02的数组作为t2
t2 = np.arange(0.0, 5.0, 0.02)

# 使用np.arange函数生成一个从0到5的步长为0.1的数组作为t1。
# 使用f(t)函数计算每个t1值的函数值。
# 使用plt.subplot(211)将整个绘图区域划分为两行一列，并将当前的子图设定为第一个子图。
plt.subplot(211)
plt.plot(t1, f(t1), "bo-", markerfacecolor="r", markersize=5)
plt.title("A tale of 2 subplots")
plt.ylabel("Damped oscillation")

plt.subplot(212)
plt.plot(t2, np.cos(2*np.pi*t2), "r--")
plt.xlabel("time (s)")
plt.ylabel("Undamped")
```

```python
def f(x):
    return np.exp(-x) * np.cos(2*np.pi*x)

# 在 matplotlib 中，plt.GridSpec() 函数用于创建一个 subplot 网格，这个网格可以用来在一个 Figure 对象中排列多个子图。
# 
# 语法：plt.GridSpec(nrows, ncols, wspace, hspace)
# 
# - nrows：grid 的行数
# - ncols：grid 的列数
# - wspace：子图之间的水平间隔
# - hspace：子图之间的垂直间隔
# 
# 该函数返回一个 GridSpec 对象，可以通过该对象的索引来选择不同的子图进行绘制。
# 
# 在给定的示例中，创建了一个 2x3 的网格，即有 2 行和 3 列。子图之间的水平间隔为 0.4，垂直间隔为 0.3。

x = np.arange(0.0, 3.0, 0.01)
grid = plt.GridSpec(2, 3, wspace=0.4, hspace=0.3)
 
# 这段代码定义了一个函数f(x)，并在一个2x3的网格中绘制了三个子图。第一个子图在第一行第一列，绘制了函数f(x)的曲线。第二个子图在第一行的剩余列，绘制了红色虚线的f(x)曲线。第三个子图在第二行的所有列，绘制了绿色的点划线的f(x)曲线。
plt.subplot(grid[0, 0])
plt.plot(x, f(x))

plt.subplot(grid[0, 1:])
plt.plot(x, f(x), "r--", lw=2)

plt.subplot(grid[1, :])
plt.plot(x, f(x), "g-.", lw=3)
```

```python
mu, sigma = 100, 15
x = mu + sigma * np.random.randn(10000)

# histtype：指定直方图的类型，可以是'bar'（默认），'barstacked'，'step'，'stepfilled'四种类型之一。在这行代码中，设置为'step'，表示绘制间断的线条。
plt.hist(x, bins=50, density=True, color="r", histtype='step')
plt.xlabel('Smarts')
plt.ylabel('Probability')
plt.title('Histogram of IQ')
plt.text(60, .025, r'$\mu=100,\ \sigma=15$')
plt.xlim(40, 160)
plt.ylim(0, 0.03)
```

```python
from scipy.stats import norm
mu, sigma = 100, 15
x = mu + sigma * np.random.randn(10000)

# 这段代码首先使用plt.hist()函数对给定的数据x进行直方图的绘制，并指定参数density=True将直方图的值归一化。函数的返回值包括直方图的统计结果、每个区间的边界值和每个区间的宽度。
# bins参数设置50个柱子
# 然后，norm.pdf()函数根据给定的均值mu和标准差sigma计算了这些边界值处正态分布的概率密度函数（PDF）的值。函数的返回值为一个与边界值对应的概率密度值的数组。
_, bins, __ = plt.hist(x, 50, density=True)
y = norm.pdf(bins, mu, sigma)
plt.plot(bins, y, 'r--', lw=3)  
plt.xlabel('Smarts')
plt.ylabel('Probability')
plt.title('Histogram of IQ')
plt.text(60, .025, r'$\mu=100,\ \sigma=15$')
plt.xlim(40, 160)
plt.ylim(0, 0.03)
```

```python
class Die():
    "模拟一个骰子的类"
    
    def __init__(self, num_sides=6):
        self.num_sides = num_sides
    
    def roll(self):
        return np.random.randint(1, self.num_sides+1)
```

* 重复投一个骰子

```python
die = Die()
results = []
for i in range(60000):
    result = die.roll()
    results.append(result)
    
plt.hist(results, bins=6, range=(0.75, 6.75), align="mid", width=0.5)
plt.xlim(0 ,7)
```

* 重复投两个骰子

```python
die1 = Die()
die2 = Die()
results = []
for i in range(60000):
    result = die1.roll()+die2.roll()
    results.append(result)
    
plt.hist(results, bins=11, range=(1.75, 12.75), align="mid", width=0.5)
plt.xlim(1 ,13)
plt.xticks(np.arange(1, 14))
```

### 13.1.6 误差图

```python
x = np.linspace(0, 10 ,50)
dy = 0.5
y = np.sin(x) + dy*np.random.randn(50)

plt.errorbar(x, y , yerr=dy, fmt="+b")

# 这段代码的作用是在网格区间[0,10]内生成50个等间距的数据点，并通过给定的偏差值dy，在每个数据点的基础上添加一个服从正态分布的随机误差。然后，使用`plt.errorbar()`函数绘制这些数据点，其中yerr参数指定了误差范围，fmt参数表示误差条及数据点显示的样式，"+b"代表蓝色加号标志。这样，绘制的图形中每个数据点都有一个垂直方向的误差条，用于表示测量误差的范围。
```

### 13.1.8 三维图形简介

【1】三维数据点与线

```python
from mpl_toolkits import mplot3d

ax = plt.axes(projection="3d")
zline = np.linspace(0, 15, 1000)
xline = np.sin(zline)
yline = np.cos(zline)
ax.plot3D(xline, yline ,zline)

zdata = 15*np.random.random(100)
xdata = np.sin(zdata)
ydata = np.cos(zdata)
ax.scatter3D(xdata, ydata ,zdata, c=zdata, cmap="spring")
```

【2】三维数据曲面图

```python
def f(x, y):
    return np.sin(np.sqrt(x**2 + y**2))

x = np.linspace(-6, 6, 30)
y = np.linspace(-6, 6, 30)
# 使用NumPy库中的"meshgrid"函数将一维数组"x"和"y"转换为二维的数组"X"和"Y"，分别表示网格中的x坐标和y坐标
X, Y = np.meshgrid(x, y)
Z = f(X, Y)

# 这行代码创建了一个3D坐标轴对象(ax)。"projection"参数指定了坐标轴的投影类型。在这个例子中，使用"3d"表示创建一个三维坐标轴。
ax = plt.axes(projection="3d")
ax.plot_surface(X, Y, Z, cmap="viridis")
# 最后，使用Matplotlib库中的"plot_surface"函数绘制一个三维表面图，其中X和Y分别表示网格中的x坐标和y坐标，Z表示对应的函数值。同时，设置了颜色映射为"viridis"来表示函数值的高低
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d

t = np.linspace(0, 2*np.pi, 1000)
X = np.sin(t)
Y = np.cos(t)
Z = np.arange(t.size)[:, np.newaxis]

ax = plt.axes(projection="3d")
ax.plot_surface(X, Y, Z, cmap="viridis")
```
